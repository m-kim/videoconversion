```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os,re

def writeVel(fn, flow):
    file = open(fn, "w")
    for col in flow:
        for tup in col:
            file.write(str(tup[0]) + ' ' + str(tup[1]) + '\n')
    file.close()
    
#cap = cv2.VideoCapture("ECEI.009141.mp4")
FILENAME = 'ECEI.009141_'
IN_DIR = '/home/mkim/Downloads/ECEI.009141/cropped_resized/'
cnt = 1
cur_fn = IN_DIR + FILENAME + str(cnt) + '.png'
frame1 = cv2.imread(cur_fn)
prvs = cv2.cvtColor(frame1,cv2.COLOR_BGR2GRAY)
hsv = np.zeros_like(frame1)

hsv[...,1] = 255
for cnt in range(2,150):
    cur_fn = IN_DIR + FILENAME + str(cnt) + '.png'
    frame2 = cv2.imread(cur_fn)#cap.read()
    next = cv2.cvtColor(frame2,cv2.COLOR_BGR2GRAY)
    
    flow = cv2.calcOpticalFlowFarneback(prvs,next, None, 0.5, 3, 15, 3, 5, 1.2, 0)
    
    mag, ang = cv2.cartToPolar(flow[...,0], flow[...,1])
    hsv[...,0] = ang*180/np.pi/2
    hsv[...,2] = cv2.normalize(mag,None,0,255,cv2.NORM_MINMAX)
    bgr = cv2.cvtColor(hsv,cv2.COLOR_HSV2BGR)
    
    #cv2.imshow('frame2',bgr)
    plt.imshow(bgr)
    cv2.imwrite('opticalfb-' + str(cnt) + '.png',frame2)
    cv2.imwrite('opticalhsv-' + str(cnt) + '.png',bgr)
    
    writeVel('XGC_' + str(cnt) + '.vel', flow)
    prvs = next


```
